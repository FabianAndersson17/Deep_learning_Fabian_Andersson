# Image processing

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# In Pyhtin script in vscode
# os.chdir(os.path.dirname(__file__))

curremt_dir = os.path.abspath("")
print(f"{curremt_dir=}")

# list all files in current dir
curremt_dir_files = os.listdir(curremt_dir)
print(f"{curremt_dir_files}")

data_dir = os.path.abspath("../Data")
print(f"{data_dir=}")

os.listdir(data_dir)
```

```python
image = plt.imread(f"{data_dir}/Homer_Simpson_2006.png")
(f"Original shape: {image.shape}")

resize_factor = .5
new_size = int(image.shape[1] * resize_factor), int(image.shape[0] * resize_factor)

image = cv2.resize(image, new_size)
plt.imshow(image)
```

## Kernals

```python
ridge_filter = np.array([[0, -1, 0],
                        [-1, 4, -1],
                        [0, -1, 0]])

blur_filter = np.array([[0.625, 0.125, 0.625],
                        [0.125, 0.25, 0.125],
                        [0.625, 0.125, 0.625]])

outline_filter = np.array([[-1, -1, -1],
                        [-1, 18, -1],
                        [-1, -1, -1]])

filters = [ridge_filter, blur_filter, outline_filter]
filter_names = ["ridge", "blur", "outline"]
```

```python
processed_images_path = f"{data_dir}/processed_images"

try:
    os.mkdir(processed_images_path)
except FileExistsError as err:
    print(err)

fig, axes = plt.subplots(1, len(filters))

for ax, filter, filter_name, in zip(axes, filters, filter_names):
    filtered_img = cv2.filter2D(image, -1, kernel=filter)
    ax.imshow(filtered_img)
    ax.axis("off")
    ax.set(title = f"{filter_name} kernal")

fig.savefig(f"{processed_images_path}/Filtered_images.png")
```
